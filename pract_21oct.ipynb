# BINARY NUMBER SYSTEM

```python
# WHY DO WE USE BINARY NUMBER SYSTEM ?
'''BECAUSE,COMPUTER ONLY UNDERSTAND ON/OFF SIGNAL AND EVERYTHING(DATA)
IS STORED IN THE FORM OF 0'S AND 1'S INCLUDING (PICTURES,NUMBERS,TEXT,IMAGES,PROGRAMS)

Decimal vs Binary
System	Base	Digits Used	Example
Decimal	10	    0-9	        123
Binary	2	    0-1	        1011

inary Value Place System

Just like decimal uses powers of 10,
Binary uses powers of 2

Binary Position	2¬≥	2¬≤	2¬π	2‚Å∞
Value	        8	4	2	1

1)CONVERT BINARY TO DECIMAL:
1011 INTO DECIMAL:
2^0*1 + 2^1*1 + 2^2*0 + 2^3*1 = 1+2+0+8 = 11(DECIMAL)

2)CONVERT DECIMAL TO BINARY
13 Divide by 2:

Step	Divide	  Quotient	Remainder
1	    13 √∑ 2	       6	1
2	    6 √∑ 2	       3	0
3	    3 √∑ 2	       1	1
4	    1 √∑ 2	       0	1


Print(bin(13))       # convert decimal ‚Üí binary
print(int("1101",2)) # convert binary ‚Üí decimal
'''
```
