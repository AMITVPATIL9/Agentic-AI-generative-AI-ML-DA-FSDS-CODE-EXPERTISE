# python program to generate otp

```python
import random

def generate_otp(length = 4):
    """ generate a numeric otp of specified length"""
    digits = '012345'
    otp = ''.join(random.choice(digits) for i in range(length))
    return otp
#Example usage
otp_length = 4
otp = generate_otp(otp_length)
print(f"your OTP is: {otp}")
```
