```python
# How to working with joint plots

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

- basically joint plot is scatter plot and it displays the relationship between audience and critic
- also if you look up you will find the uniform distribution(critics) and normal distribution(audience)
