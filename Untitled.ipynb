```python
x = 10  # Global variable

def update_x():
    global x  # Declare that we are using the global variable x
    x += 5    # Modify the global variable
   
update_x()
print(x)  # Output: 15
```

```python
def count():
    list1 = []
    even = 0
    odd = 0
    print('enter numbers')

    list1 = list(map(int, input().split()))
    for i in list1:
        if i%2 == 0:
            even += 1
        else:
            odd += 1
    print('no of evens are:',even)
            
    
count()    
```

```python
'''
✅ What’s Causing the Error?
1. input(int()) is incorrect

input() takes no arguments.
input(int()) evaluates int() → 0, so it becomes input(0) which Python interprets as a format string → TypeError.

2. You're treating input as a list, but it's a string

list = input() returns a string, so iterating over it gives characters, not numbers.

3. odd is not defined

You increment odd but never initialized it.

4. Avoid using list as a variable name

It shadows the built-in list type'''
```
