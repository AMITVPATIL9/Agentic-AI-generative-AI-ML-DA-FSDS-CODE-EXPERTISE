# superset,subset,disjoint

```python
s4 = {1,2,3,4,5,6,7,8,9}
s5 = {3,4,5,6,7,8}
s6 = {10,20,30,40}
```

```python
s4.issuperset(s5) # dono mein se ek hona chaahiye
```

```python
s5.issuperset(s6)
```

```python
s4.issuperset(s5)
```

```python
s6.isdisjoint(s4)  # disjoint no common element
```

```python
s7 = {1,2,3,4,5,6,7,8,9,15}
s8 = {15,25,35}
s9 = {10,20,30,40}
```

```python
s8.issubset(s7) # subset matlab saare element hone chaiye usme
```

```python
s8.isdisjoint(s7)
```
